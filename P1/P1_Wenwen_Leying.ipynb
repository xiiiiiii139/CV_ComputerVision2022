```python
# create an RGB black image that has the same shape as Lenna image and print it 
lenna_length = img3.shape[0]
lenna_width = img3.shape[1]
black_img = np.zeros((lenna_length,lenna_width,3),dtype = np.uint8)
plt.axis('off')
print('Original Lenna shape:',img3.shape)

# convert the image 10% larger and print the shape of the frame which is 10% larger
from skimage.transform import rescale
black_bk = rescale(black_img,1.1)
plt.imshow(black_bk)
plt.axis('off')
print('Black background shape:',black_bk.shape)

background_length = black_bk.shape[0]
background_width = black_bk.shape[1]

# in order to center de image, we should calculate the coordinates where the first pixel of the Lenna will start
y = round((background_length-lenna_length)/2)
x = round((background_width-lenna_width)/2)

# so the first pixel where we should start will be:
# print('first pixel:',(y,x))

# use numpy indexing to embed Lenna image in the center of background image
result = black_bk.copy()
result[y:y+lenna_length, x:x+lenna_width] = img3

# rescaled images are saved with float64 data type, so we need to convert them in int to show it correctly
result = result.astype(int)

# plot result
plt.imshow(result)
plt.title('Embed Lenna')
```

b) Change the frame of the new image to green. Visualize the three images: original, black frame and green frame, in one single figure, with their respective titles.

```python
"""
function that returns the image with the frame of specified color
    r: channel red 0-255
    g: channel green 0-255
    b: channel blue 0-255
    percentLarger: the percentage we want to be longer
    image: image we want to show
    
    returns:
    image with the frame of specified color
"""

def frame(r, g, b, percentLarger, image):
    image_length = image.shape[0]
    image_width = image.shape[1]
    
    #create a new frame
    new_img = np.zeros((image_length,image_width,3),dtype = np.uint8)
    
    #convert the image % larger
    from skimage.transform import rescale
    scale = 1+(percentLarger/100)
    new_bk = rescale(new_img, scale)
    
    new_bk[:,:,0] = r
    new_bk[:,:,1] = g
    new_bk[:,:,2] = b
    
    background_length = new_bk.shape[0]
    background_width = new_bk.shape[1]
    
    # in order to center de image, we should calculate the coordinates where the first pixel of the Lenna will start
    y = round((background_length-image_length)/2) 
    x = round((background_width-image_width)/2)
    
    # so the first pixel where we should start will be:
    #print('first pixel:',(y,x))

    # use numpy indexing to embed Lenna image in the center of background image
    result = new_bk.copy()
    result[y:y+image_length, x:x+image_width] = image
    
    # rescaled images are saved with float64 data type, so we need to convert them in int to show it correctly
    result = result.astype(int)

    return result
```

```python
canvas, axs = plt.subplots(nrows=1, ncols=3, figsize=(12,12))

# the original image
ax = axs[0]
ax.imshow(img3)
ax.axis('off')
ax.set_title('Original image')

# the black frame image
ax = axs[1]
ax.imshow(result,cmap='gray') # plot image from the previous exercise
ax.axis('off')
ax.set_title('Black frame image')

# the green frame image
ax = axs[2]
# Note: The RGB color value of green is (0,255,0)
gren_frame = frame(0,255,0,10,img3) # we use the function decribed previously 
ax.imshow(gren_frame)
ax.axis('off')
ax.set_title('Green frame image')
```

d) Apply contrast enhancement to the original image. There are two ways for it: (1) by direct manipulation, and (2) through scikit-image command [rescale_intensity](https://scikit-image.org/docs/stable/api/skimage.exposure.html#skimage.exposure.rescale_intensity). Check the minimal and maximal values of the enhanced image, and compare them with the previous one. Show the two images  without graylevel enhancement.

```python
# Bvout = (Bvin-mink/maxk-mink)*quantk
from skimage.util import img_as_ubyte

def direct_manipulation(image,quantk):
    """
    function that apply contrast enhancement through direct manipulation
    """
    mink = image.min()
    maxk = image.max()
    direct = image.copy()
    direct = ((direct-mink)/(maxk-mink))*quantk
    
    return direct
```

```python
# direct manipulation
direct =  direct_manipulation(img5,255)

# sickit image
from skimage import exposure
sickit = exposure.rescale_intensity(img5)

print('Details of image through direct manipulation')
show_details(direct)
print()

print('Details of image through sickit')
show_details(sickit)
canvas, axs = plt.subplots(nrows=1,ncols=2,figsize=(12,12))

# direct manipulation image
ax = axs[0]
ax.imshow(direct,cmap='gray',vmin = 0,vmax = 255)
ax.axis('off')
ax.set_title('Direct manipulation image')

# sickit image
ax = axs[1]
ax.imshow(sickit,cmap='gray', vmin = 0, vmax = 255)
ax.axis('off')
ax.set_title('Sickit image')
```

```python
"""
function which capture the three diferent color channels
    color: color channel
    image: image we want to modify

    returns:
    image modified
"""
def channel(color, image):
    new_img = image.copy()
    if color == 'red':
        new_img[:,:,1] = 0
        new_img[:,:,2] = 0
    elif color == 'green':
        new_img[:,:,0] = 0
        new_img[:,:,2] = 0
    else:
        new_img[:,:,0] = 0
        new_img[:,:,1] = 0
    
    return new_img
```

b) Given the images `images_notebook/coat.png` and `images_notebook/model.png` (first two images on the figure below), implement the necessary code to construct the image of the model with the coat (third image). Next find in internet an image with a texture and apply it to obtain an image of the jacket with the found texture (fourth image is only an example).

<img src="images_notebook/ex17d.png">

**Note:** when applying [resize, rotate, etc...](https://scikit-image.org/docs/stable/auto_examples/transform/plot_rescale.html) keep track how the image type and values range change. When necessary, apply type conversion using [img_as_float, img_as_uint, img_as_ubyte.](https://scikit-image.org/docs/stable/api/skimage.html)

**Note: DO NOT USE LOOPS to overlap the images!! It will be penalized.**


```python
"""
function that changes the texture of the coat and we put it on the model
    model: the image of the model that we have to put clothes on
    clothing: the image with the shape of clothing
    texture: image with the texture of the clothes
"""
def changeTexture(model, clothing, texture):
    # we save the shape of the original image
    length = model.shape[0]
    width = model.shape[1]
    
    # resize the new texture image if is not the same shape as the others
    if texture.shape[0] != length or texture.shape[1] != width:
        texture = resize(texture,(length,width))
    
    from skimage.color import rgba2rgb
    # convert rgba to rgb
    model_rgb = rgba2rgb(model)
    cloth_rgb = rgba2rgb(clothing)
    # in case that the texture is rgba, we need to convert it to rgb too
    texture_rgb = texture.copy()
    if texture.shape[2] > 3:
        texture_rgb = rgba2rgb(texture)
    
    # apply type conversion
    model_rgb_int = img_as_ubyte(model_rgb)
    cloth_rgb_int = img_as_ubyte(cloth_rgb)
    texture_rgb_int = img_as_ubyte(texture_rgb)
    
    # all the values of pixels that are not black in coat, will be new model pixels
    # we create a new image which have the same shape as the original image, this one will be returned
    # model with texture coat
    newImage = model_rgb_int.copy()
    newImage[cloth_rgb_int != 0] = texture_rgb_int[cloth_rgb_int != 0]
    
    return newImage
```

```python
from skimage.util import img_as_ubyte
from skimage.transform import resize

# load image from images_notebook
coat = io.imread('images_notebook/coat.png')
model = io.imread('images_notebook/model.png')
texture = io.imread('images_notebook/texture2.png')
# a new texture found in internet
wall_texture = io.imread('images_notebook/wall_texture.jpg')

# we use the function to create the new texture clothing images
yellowCoat = changeTexture(model, coat, coat)
redCoat = changeTexture(model, coat, texture)
newCoat = changeTexture(model, coat, wall_texture)

# now we plot the resolved images
canvas, axs = plt.subplots(nrows=1, ncols=3, figsize=(12,12))

# model with yellow coat
ax = axs[0]
ax.imshow(yellowCoat, cmap='gray',vmin = 0, vmax = 1)
ax.axis('off')
ax.set_title('Model with yellow coat')

# model with texture coat
ax = axs[1]
ax.imshow(redCoat, cmap='gray',vmin = 0, vmax = 1)
ax.axis('off')
ax.set_title('Model with texture coat')

# model with new texture coat
ax = axs[2]
ax.imshow(newCoat, cmap='gray',vmin = 0, vmax = 1)
ax.axis('off')
ax.set_title('Model with new texture coat')
```
